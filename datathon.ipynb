```python
import pandas as pd
import numpy as np
data = pd.read_csv(r'train.csv', index_col = 0)
```

```python
data.isna()
data.dropna(inplace = True)
data_1 = data.set_index("MSN")
data_new = data_1.iloc[:,1:]
data_new
```

```python
data_sub = data_new.iloc[:29,:]
data_sub2 = data_sub.transpose()
data_sub2
```

```python
df = pd.DataFrame()
for place in set(data_new['State']):
    for time in set(data_new['Year']):
        data_sub = data_new.loc[(data_new.State == place) & (data_new.Year == time)]
        data_sub2 = data_sub.transpose()
        data_sub3 = data_sub2.loc['Amount',:]
        df = pd.concat([df,data_sub3], axis = 1)
df
        
```

```python
df = pd.DataFrame()
state_df = pd.DataFrame()
for place in set(data_new['State']):
    state = {'State': place}
#     print(state)
    for time in set(data_new['Year']):
        data_sub = data_new.loc[(data_new.State == place) & (data_new.Year == time)]
        data_sub2 = data_sub.transpose()
        data_sub3 = data_sub2.loc['Amount',:]
        df = pd.concat([df,data_sub3], axis = 1)
        state_df = state_df.append(state, ignore_index = True)
        #print(state_df)
df = df.transpose()
df.index = [i for i in range(250)]
df = pd.concat([df, state_df],axis=1)
```

```python
df = pd.DataFrame()
state_df = pd.DataFrame()
co2_df = pd.DataFrame()
inves_df = pd.DataFrame()

for place in set(data_new['State']):
    state = {'State': place}
    co2 = {'CO2': data_new[data_new['State']==place]['CO2 Emissions (Mmt)'][0]}
    inves = {'TotalAmountofAssistance': data_new[data_new['State']==place]['TotalAmountofAssistance'][0]}
    for time in set(data_new['Year']):
        data_sub = data_new.loc[(data_new.State == place) & (data_new.Year == time)]
        data_sub2 = data_sub.transpose()
        data_sub3 = data_sub2.loc['Amount',:]
        df = pd.concat([df,data_sub3], axis = 1)
        state_df = state_df.append(state, ignore_index = True)
        co2_df = co2_df.append(co2, ignore_index = True)
        inves_df = inves_df.append(inves, ignore_index = True)
        #print(state_df)
df = df.transpose()
df.index = [i for i in range(250)]
df = pd.concat([df, state_df],axis=1)
df = pd.concat([df, co2_df],axis=1)
df = pd.concat([df, inves_df],axis=1)
```

```python
from sklearn import linear_model
x_lst = list(df.columns[:-1])

x = df[x_lst]
y = df['TotalAmountofAssistance']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(mean_squared_error(y_test, y_pred)) 
print(model.coef_) #权重
print(model.n_features_in_) #几个输入
```
